```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_excel(r'Moving Data.xlsx', sheet_name='Moving Data')
```

```python
east = df['UTM_Easting (in m)']
north = df['UTM_Northing (in m)']
alt = df['Altitude (in m)']
time = df['Time (in seconds)']
```

```python
plt.scatter(east, north)
plt.rcParams.update({'figure.figsize':(10,8), 'figure.dpi':100})
plt.title('UTMEasting vs UTMNorthing')
plt.xlabel('UTMEasting')
plt.ylabel('UTMNorthing')
sns.regplot(x=east, y=north, lowess=True, color = "blue")
plt.show()
```

```python
plt.scatter(time,alt)
#plt.rcParams.update({'figure.figsize':(10,8), 'figure.dpi':100})
plt.title('Altitude vs Time')
plt.xlabel('Time')
plt.ylabel('Altitude')
sns.regplot(x=time, y=alt, lowess=True, color = "red")
plt.show()
```

```python
def rsme (column_name):
 column = np.mean(column_name)
 column_rsme = np.sqrt(np.square(column_name - column).mean())
 return column_rsme
```

```python
rsme(east)
```

```python
rsme(north)
```

```python
def error_bounds (column_name):
 column=df[column_name].tolist()
 mean = np.mean(column) 
 u = max(column) - np.mean(column)
 l = np.mean(column)  - min(column)
 print(f'upper bound : {u}')
 print(f'mean: {mean}')
 print(f'lower bound : {l}')
```

```python
error_bounds('UTM_Easting (in m)')
```

```python
error_bounds('UTM_Northing (in m)')
```
